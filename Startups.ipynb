```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={'City':['San Francisco Bay', 'New York', 'Beijing', 'Los Angeles Area', 'London', 'Boston Area', 'Shanghai', 'Tel Aviv Area', 'Moscow', 'Bangalore'],
      'Quality Score': [296.02,95.55,58.61,43.41,37.44,40.53,36.17,19.92,15.64,17.95],
     }
df=pd.DataFrame(data)
df
```

### Line Chart for Top 10 Cities Quality Score 

```python
df.plot(kind='line')

plt.xticks(rotation=45, fontsize=17)
plt.yticks(rotation=45, fontsize=17)

plt.xlabel('Top 10 Cities', fontsize=17)
plt.ylabel('Quality Score', fontsize=17)

plt.title('Top 10 Cities Quality Score', fontsize=17)
plt.show()
```
